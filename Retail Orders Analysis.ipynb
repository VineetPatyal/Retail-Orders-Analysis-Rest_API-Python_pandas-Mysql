```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() # extract file to directory
zip_ref.close() #close file
```

```python
#2 Read data from the file and handle null values
import pandas as pd
```

```python
df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

```python
df.head(10)
```

```python
#3 rename columns names, make them lower case and repalce space with underscore
```

```python
df.columns=df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_' )
```

```python
df.head(10)
```

```python
#4 Derive new columns discount, sale price and profit
```

```python
df['discount']= df['list_price']*df['discount_percent']*.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
```

```python
df.head()
```

```python
#5 convert order date from object data type to datetime
df['order_date']=pd.to_datetime(df['order_date'],format= "%Y-%m-%d")
df.dtypes
```

```python
#6 drop cost price,list price & discount percent columns.
df.drop(columns=['list_price','cost_price','discount_percent'],inplace = True)
```
